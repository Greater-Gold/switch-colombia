```python
import pandas as pd
model_inputs_path = '../../../model/inputs/'
timepoints = pd.read_csv(model_inputs_path+'timepoints.csv')
timepoints['timepoints'] = timepoints['timepoint_id']
# Eliminar columnas innecesarias
timepoints = timepoints.drop(columns=['timeseries','timepoint_id'])

model_outputs_path = '../../../model/outputs/'
dispatch_wide = pd.read_csv(model_outputs_path+'dispatch_wide.csv')
# Añadir timepoints
dispatch_wide = pd.merge(dispatch_wide, timepoints, on='timestamp', how='inner')
# Eliminar columnas innecesarias
dispatch_wide = dispatch_wide.drop(columns=['timestamp'])
dispatch_wide.head()
```

```python
import pandas as pd
import plotly.express as px

columns_to_plot = [col for col in dispatch_wide.columns if col.startswith('S_')]

# Pivotar el DataFrame para que sea adecuado para plotly express
df_melt = dispatch_wide.melt(id_vars=['timepoints'], value_vars=columns_to_plot, 
                  var_name='Generador', value_name='Capacidad')
df_melt.head()

# Utilizando plotly.express para crear el gráfico
fig = px.line(df_melt, x='timepoints', y='Capacidad', color='Generador', 
              title='Producción por Generador a lo largo del Tiempo', markers=True)

# Mostrar el gráfico
fig.show()

```
